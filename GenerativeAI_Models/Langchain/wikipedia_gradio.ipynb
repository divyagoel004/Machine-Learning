```python
from langchain.agents import load_tools
from langchain.agents import initialize_agent
from langchain.agents import AgentType
from together import Together
from langchain_openai import ChatOpenAI
import requests
from bs4 import BeautifulSoup
import os

os.environ['TOGETHER_API_KEY'] = "666e8dc46b45912071558831227d603654035a8f70c3e9f0a41f6f0f9661ad98"
```

```python


model = ChatOpenAI(
    base_url="https://api.together.xyz/v1",
    api_key=os.environ["TOGETHER_API_KEY"],
    model="mistralai/Mixtral-8x7B-Instruct-v0.1",
)
```

```python
tool = load_tools(
    ['wikipedia'],
    api_key=serpapi,
    model=model,
)
```

```python
agent=initialize_agent(tool,model,agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,verbose=True)
```

```python
agent.run("Which journal released the paper named Attention is all you need?")
```

```python
import gradio as gr

# Define a function to use the Wikipedia agent
def query_wikipedia(query):
    try:
        # The agent takes a query and returns a result
        result = agent.run(query)
        return result
    except Exception as e:
        return str(e)

# Define the Gradio interface
# Create a text input for the user query and text output for the response
interface = gr.Interface(
    fn=query_wikipedia,
    inputs="text",
    outputs="text",
    title="Wikipedia Agent with Gradio",
    description="Ask any question, and the Langchain agent will fetch information from Wikipedia."
)

# Launch the Gradio app
interface.launch()
```
